### Vector Operations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vector1= np.array([1,2,3,4,5])
vector2= np.array([6,7,8,9,0])

print("Vector addition", vector1+ vector2)
print("Vector mutliplication", vector1* vector2)
print("DOt Product", np.dot(vector1,vector2))

angle  = np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)* np.linalg.norm(vector2)))
print(angle)
```

## Simpe demo 

```python
prices = np.linspace(5, 30, 20)
base_demand = 1000 * np.exp(-0.1* prices)
sales = base_demand + np.random.normal(0,10,20)
revenue = prices *sales
optimal_index = np.argmax(revenue)


plt.figure(figsize=(8,4))
plt.plot(prices, sales , 'b-', label= "Sales")
plt.plot(prices, revenue, 'r-', label = "Revenue")
plt.axvline(prices[optimal_index], color="g", linestyle="--",label = f"optimal price")
plt.legend()
plt.title("Price Senesitivity analysis")
plt.xlabel('Price(Rs)')
plt.show()
```
