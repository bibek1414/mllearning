```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data
mu, sigma = 0, 1  # Mean and standard deviation
data = np.random.normal(mu, sigma, 1000)

# Plot the histogram and probability density function
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
plt.plot(x, norm.pdf(x, mu, sigma), 'r', linewidth=2)
plt.title('Normal Distribution')
plt.show()
```

```python
from scipy.stats import binom

# Prior probabilities
p_fair = 0.5  # Probability the coin is fair
p_biased = 0.5  # Probability the coin is biased

# Likelihoods (assuming we see 8 heads out of 10 flips)
likelihood_fair = binom.pmf(8, 10, 0.5)  # Fair coin (50% heads)
likelihood_biased = binom.pmf(8, 10, 0.7)  # Biased coin (70% heads)

# Posterior probability using Bayes' Theorem
p_heads = likelihood_fair * p_fair + likelihood_biased * p_biased
posterior_fair = (likelihood_fair * p_fair) / p_heads
posterior_biased = (likelihood_biased * p_biased) / p_heads

print(f"Updated Probability of Fair Coin: {posterior_fair:.3f}")
print(f"Updated Probability of Biased Coin: {posterior_biased:.3f}")
```

```python
from scipy.optimize import minimize

# Sample data
data = np.random.normal(5, 2, 100)  # True mean=5, std=2
print(data)
# Define negative log-likelihood function
def negative_log_likelihood(mu):
    return -np.sum(np.log(norm.pdf(data, mu, 2)))

# Find the MLE for mean
result = minimize(negative_log_likelihood, x0=np.mean(data))
estimated_mu = result.x[0]

print(f"Estimated Mean (MLE): {estimated_mu:.3f}")
```
