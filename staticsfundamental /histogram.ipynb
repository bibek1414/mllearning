# Histogram 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load a sample dataset
from sklearn.datasets import load_iris
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Plot histogram of petal length
plt.hist(df['petal length (cm)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

# 

```python
import numpy as np

# Original vs log-transformed histogram
plt.subplot(1, 2, 1)
plt.hist(df['petal width (cm)'], bins=20)
plt.title("Original")

plt.subplot(1, 2, 2)
plt.hist(np.log1p(df['petal width (cm)']), bins=20)
plt.title("Log Transformed")

plt.tight_layout()
plt.show()
```
