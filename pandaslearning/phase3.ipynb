```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Sample DataFrame
df = pd.DataFrame({
    'A': np.random.randn(100).cumsum(),
    'B': np.random.randn(100).cumsum(),
    'C': np.random.randn(100).cumsum()
})

# Line plot
df.plot()
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()

# Bar plot
df.iloc[0:10].plot.bar()
plt.title('Bar Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()

# Stacked bar plot
df.iloc[0:10].plot.bar(stacked=True)
plt.title('Stacked Bar Plot')
plt.show()

# Horizontal bar plot
df.iloc[0:10].plot.barh()
plt.title('Horizontal Bar Plot')
plt.show()

# Histogram
df.plot.hist(bins=20, alpha=0.5)
plt.title('Histogram')
plt.show()

# Box plot
df.plot.box()
plt.title('Box Plot')
plt.show()

# Area plot
df.plot.area(alpha=0.5)
plt.title('Area Plot')
plt.show()

# Scatter plot
df.plot.scatter(x='A', y='B')
plt.title('Scatter Plot')
plt.show()

# Pie chart
abs(df.iloc[0]).plot.pie(autopct='%.2f%%')
plt.title('Pie Chart')
plt.show()
```
