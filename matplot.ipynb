# Matplot lib

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]  # y = x²

# Make the plot
plt.figure(figsize=(8, 5))  # Set the size (width, height) in inches
plt.plot(x, y)
plt.title("Square Function")
plt.xlabel("x")
plt.ylabel("y = x²")
plt.grid(True)  # Add a grid
plt.show()
```
