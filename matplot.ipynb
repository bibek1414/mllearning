# Matplot lib

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]  # y = x²

# Make the plot
plt.figure(figsize=(8, 5))  # Set the size (width, height) in inches
plt.plot(x, y)
plt.title("Square Function")
plt.xlabel("x")
plt.ylabel("y = x²")
plt.grid(True)  # Add a grid
# More customization options
plt.plot(x, y, color='red', linestyle='--', marker='o', linewidth=3, markersize=8)
plt.xlim(0, 6)  # Set x-axis limits
plt.ylim(0, 30)  # Set y-axis limits
plt.show()
# Save your plot as a file
plt.savefig('my_plot.png')  # Save as PNG image
plt.savefig('my_plot.pdf')  # Save as PDF (good for papers)
```
